# Task
Analyze the sales data in "superstore.csv" to calculate and display the following metrics: total sales, sales growth percentage compared to last week, top 3 products by sales, and top 3 employees by sales.

## Calculate total sales

### Subtask:
Calculate the sum of the 'Sales' column in the DataFrame.


**Reasoning**:
Calculate the sum of the 'Sales' column and store it in the `total_sales` variable.



```python
total_sales = df['Sales'].sum()
print(f"Total Sales: {total_sales}")
```

## Calculate sales growth percentage (compared to last week)

### Subtask:
Calculate sales growth percentage (compared to last week).


**Reasoning**:
Convert 'Order Date' to datetime, calculate sales for the last two weeks, and compute sales growth percentage.



```python
# Convert 'Order Date' to datetime objects (already done in previous cell but keeping for completeness)
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Determine the last week and the week before the last week
max_date = df['Order Date'].max()
last_week_start = max_date - pd.Timedelta(days=6)
week_before_last_week_start = max_date - pd.Timedelta(days=13)

# Filter data for the last week and the week before the last week
sales_last_week_df = df[(df['Order Date'] >= last_week_start) & (df['Order Date'] <= max_date)]
sales_week_before_last_week_df = df[(df['Order Date'] >= week_before_last_week_start) & (df['Order Date'] < last_week_start)]

# Calculate the sum of 'Sales' for each week
sales_last_week = sales_last_week_df['Sales'].sum()
sales_week_before_last_week = sales_week_before_last_week_df['Sales'].sum()

# Calculate sales growth percentage, handling division by zero
if sales_week_before_last_week != 0:
    sales_growth_percentage = ((sales_last_week - sales_week_before_last_week) / sales_week_before_last_week) * 100
else:
    sales_growth_percentage = float('inf') if sales_last_week > 0 else 0

print(f"Sales last week: {sales_last_week:.2f}")
print(f"Sales week before last week: {sales_week_before_last_week:.2f}")
print(f"Sales Growth Percentage (compared to last week): {sales_growth_percentage:.2f}%")
```

## Identify top 3 products by sales

### Subtask:
Group the DataFrame by 'Product Name', calculate the sum of 'Sales' for each product, sort the products by sales in descending order, and select the top 3.


**Reasoning**:
Group the DataFrame by 'Product Name', calculate the sum of 'Sales' for each product, sort in descending order, and select the top 3 products.



```python
top_3_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(3)
display(top_3_products)
```

## Identify top 3 employees by sales

### Subtask:
Group the DataFrame by 'Customer Name', calculate the sum of 'Sales' for each customer, sort the customers by sales in descending order, and select the top 3.


**Reasoning**:
Group the DataFrame by 'Customer Name', calculate the sum of 'Sales', sort the results, select the top 3, and display them.



```python
top_3_employees = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(3)
display(top_3_employees)
```

## Display metrics

### Subtask:
Print the calculated metrics in a formatted way.


**Reasoning**:
Print the calculated metrics in a formatted way as requested.



```python
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Sales Growth Percentage (compared to last week): {sales_growth_percentage:,.2f}%")
print("\nTop 3 Products by Sales:")
display(top_3_products)
print("\nTop 3 Employees by Sales:")
display(top_3_employees)
```

## Write a function to send email

### Subtask:
Create a Python function that takes the email details and sends the email using the configured SMTP settings.


**Reasoning**:
Import necessary libraries and define the function to send the email.



```python
import smtplib
from email.mime.text import MIMEText

def send_sales_report_email(recipient_email, subject, html_body):
    """Sends a weekly sales report email with HTML content.

    Args:
        recipient_email: The email address of the recipient.
        subject: The subject of the email.
        html_body: The HTML content of the email body.
    """
    msg = MIMEText(html_body, 'html')
    msg['Subject'] = subject
    msg['From'] = sender_email
    msg['To'] = recipient_email

    try:
        with smtplib.SMTP(smtp_server, smtp_port) as server:
            server.starttls()  # Secure the connection
            server.login(sender_email, sender_password)
            server.sendmail(sender_email, recipient_email, msg.as_string())
        print("Email sent successfully!")
    except Exception as e:
        print(f"Failed to send email: {e}")
```
